```python
# carregando o dataset com dados de aluguéis
df = pd.read_csv('https://s3-sa-east-1.amazonaws.com/lcpi/164c4740-feb8-4bb2-8595-6423754567fa.csv', sep=';')
df.head()
```

```python
# selecionando 20 amostras do conjunto de dados
amostras = df.sample(n=20, random_state=101)
amostras
```

## **Métodos de estilização**

1. `highlight_min` | `highlight_max`

```python
amostras.style.highlight_min()
```

```python
amostras.style.highlight_min(subset=['Valor'], color='red')
```

```python
amostras.style.highlight_min(subset=['Valor'], props='background: yellow; color: black')
```

```python
amostras.style.highlight_max(color='green')
```

```python
amostras.style.highlight_min(subset=['Valor', 'IPTU'], 
                             props='background: green; color: black').highlight_max(subset=['Valor', 'IPTU'], 
                                                                                     props='background: red; color: black')
```

2. `set_caption`
 - adicionando legenda ao DataFrame

```python
amostras.style.set_caption('Imovéis para alugar')
```

```python
amostras.style.set_caption('Imovéis para alugar').set_table_styles(
    [{'selector': 'caption', 'props': 'font-size: 20px; font-weight: bold'}]
)
```

```python
amostras.style.set_caption('Imovéis para alugar').set_table_styles(
    [{'selector': 'caption', 'props': 'font-size: 20px; font-weight: bold; text-align: left'}]
)
```

3. `format`
 - formatando os dados das colunas

Apenas alterando a forma de apresentar os dados das colunas 👀

```python
amostras.style.format(
    {'Valor': '{:.2f}'}
)
```

```python
amostras.style.format(
    {'Valor': '{:.2f}', 'Valor m2': '{:.1f}'}
)
```

4. `applymap`
 - definindo condições para aplicar estilos

```python
personal_finances = pd.DataFrame(
    {'Description': ['food', 'salary', 'investiment', 'market'], 
     'Value': [-1100, 4000, 500, -360]}
)

personal_finances
```

```python
def apply_style(line):
    if (line['Value'] < 0):
        return ['background-color: red'] * len(line)
    else:
        return ['background-color: blue'] * len(line)
```

```python
personal_finances.style.apply(axis=1, func=apply_style)
```

```python
# salvando em um arquivo excel
personal_finances.style.apply(axis=1, func=apply_style).to_excel('test.xlsx')
```
